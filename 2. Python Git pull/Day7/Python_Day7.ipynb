Attributes:

  -ndim : return number of dimension in array

  -shape : returns rows and column as a tuple

  -size : returns total number of elements
  
  -dtype : return datatype of elements

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)
```

Aggregation Methods:



1.   sum()
2.   max()
3.   min()
4.  mean()
5.  median()
6.  std()



```python
arr = np.array([10,20,30,40,50])
print(np.sum(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.min(arr))
print(np.max(arr))
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print()
print(np.sum(arr, axis = 0))
print(np.sum(arr, axis = 1))
print(np.sum(arr))

print(arr.sum(0))
```

Universal Function:


1.   log
2.   sqrt
3.   exp
4.   sin...



```python
arr = np.array([1,4,9,16,25])
print(np.sqrt(arr))
print(np.exp(arr))
print(np.log(arr))
```

```python
angles = np.array([0, np.pi/2, np.pi])
print(np.sin(angles))
print(np.cos(angles))
```

```python
#Masking
arr = np.array([10,15,20,25,30,35])

mask = arr > 20
print(mask)

print(arr[mask])
```

```python
#Filtering
arr = np.array([10,15,20,25,30,35])

res = arr[(arr > 10) & (arr % 2 == 0)]
print(res)
```

```python
arr = np.array([1,34,67,23,89,2,48])
print(np.sort(arr))

print(np.argsort(arr))
print(np.argsort(arr[::-1]))
```
