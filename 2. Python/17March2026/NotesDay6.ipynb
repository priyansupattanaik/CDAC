````python
#@property and @setter
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    @property
    def salary(self):
        return self.__salary
    
    @salary.setter
    def salary(self, amount):
        if amount > 0:
            self.__salary = amount
        else:
            print("Invalid salary amount")

    def display(self):
        print(self.__name)
        print(self.__salary)    

emp = Employee("John Doe", 50000)
emp.display()

print("Current salary:", emp.salary)

emp.salary = 70000
print("Updated Salary:", emp.salary)
````

    def attribute_name(self):

Abstract Class VS Concrete Class

Abstract Class: 
- A class that cannot be instantiated and is meant to be subclassed. It can contain abstract methods (methods without implementation) that must be implemented by any concrete subclass.
- It serves as a blueprint for other classes and can provide common functionality to its subclasses.
- It is defined using the `abc` module in Python and typically includes the `@abstractmethod` decorator to indicate which methods are abstract.
- Example:
```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass
```

Concrete Class: A class that can be instantiated and provides implementations for all of its methods, including those inherited from abstract classes.
- It can be instantiated to create objects and can be used directly in the program.
- It may inherit from an abstract class and must implement all abstract methods defined in the abstract class
- Example:
```python
class Dog(Animal):
    def make_sound(self):
        return "Woof!"
```



````python
from abc import ABC, abstractmethod

class A(ABC):
    @abstractmethod
    def m1(self):
        pass

    def m2(self):
        print("m2 (): Normal Method")

class B(A):
    def m1(self):
        print("m1 (): implemented and executed")

a = B()
a.m1()
a.m2()
# b = A() #Error: cannot create object of abstract class
# b.m1() #Error: cannot call abstract method
# b.m2() #Error: cannot call method of abstract class
````

#Inheritance

- Inheritance is an OOP concept where one class (child class) acquires the properties and behaviors (attributes and methods) of another class (parent class).
- Parent Class / Base Class / Super Class
- Child Class / Base Class / Sub Class

- Reak World Entity
    - Ex:
    parent class: Vehicle
    child class: Car, Bike, Truck

- Syntax:
class Parentclassname:
    state1

class Childclass(Parentclassname):
    state2

-Advantages:
--------------
- code reuse
- reduce code duplication
- easy maintenance
- logical program structure 
- supports hirarchical relationship

Types of Inheritance:
-----------------------
1. Single Inheritance
2. Multiple Inheritance
3. Multilevel Inheritance
4. Hierarchical Inheritance
5. Hybrid Inheritance

1. Single Inheritance:
------------------------
- A child class inherits properties and methods from only one parent class.
- Child class can access the methods of parent class.
- This is the simplest form of inheritance.

````python
#Insingle Inheritance

class A:

    def m1(self):
        print("m1 (): parent class method")

class B(A):

    def m2(self):
        print("m2 (): child class method")

s = B() #object of child class can access both parent and child class methods
s.m1() #inherited method from parent class
s.m2() #child class method

s = A() #object of parent class can access only parent class methods
s.m1() #parent class method
# s.m2() #Error: AttributeError: 'A' object has no attribute 'm2()'
````

2. Multiple Inheritance:
------------------------
- A child class inherits properties and methods from more than one parent class.
- Child class can access members of all parent classes.
- Multiple inheritance 

````python
#Inmultiple Inheritance:

class A:
    def m1(self):
        print("m1 (): parent class 1")

class B:
    def m2(self):
        print("m2 () : parent class 2")

class C(A,B):
    def m3(self):
        print("m3 (): child class method")

c = C()
c.m1()
c.m2()
c.m3()

c = B() #object of parent class 2
#c.m1() #Error: AttributeError: 'B' object has no attribute 'm1()' 
c.m2() #parent class 2 method
#c.m3() #Error: AttributeError: 'B' object has no attribute 'm3()'

c = A() #object of parent class 1
c.m1() #parent class 1 method
#c.m2() #Error: AttributeError: 'A' object has no attribute 'm2()'
#c.m3() #Error: AttributeError: 'A' object has no attribute 'm3()'
````

3. Multilevel Inheritance:
------------------------
- Inheritance occurs in multiple levels.
- In multiple inheritance one class inherits from another derived class.

````python
#3. Multilevel Inheritance:
class A:
    def m1(self):
        print("m1 (): parent class 1")
class B(A):
    def m2(self):
        print("m2 (): parent class 2")
class C(B):
    def m3(self):
        print("m3 (): child class method")

c1 = C()
c1.m1() #inherited from class A
c1.m2() #inherited from class B
c1.m3() #own method of class C

#because we have created object of child class c, we can access all methods of parent classes A and B, as well as its own method m3().

c1 = B() #object of parent class B
c1.m1() #inherited from class A
c1.m2() #own method of class B
#c1.m3() #Error: AttributeError: 'B' object has no attribute 'm3()'

c1 = A() #object of parent class A
c1.m1() #own method of class A
#c1.m2() #Error: AttributeError: 'A' object has no attribute 'm2()'
#c1.m3() #Error: AttributeError: 'A' object has no attribute 'm3()'
````

4. Hierarchical Inheritance:
------------------------
- Multiple child classes inherit properties and methods from one parent class.
- 

````python
#hierarchical Inheritance:

class A:
    def m1(self):
        print("m1 (): parent class method")
class B(A):
    def m2(self):
        print("m2 (): child class 1 method")
class C(A):
    def m3(self):
        print("m3 (): child class 2 method")

ch1 = B()
ch1.m1() #inherited from parent class A
ch1.m2() #own method of child class B
#ch1.m3() #Error: AttributeError: 'B' object has no attribute 'm3()'

ch1 = C()
ch1.m1() #inherited from parent class A
#ch1.m2() #Error: AttributeError: 'C' object has no attribute 'm2()'
ch1.m3() #own method of child class C

ch1 = A()
ch1.m1() #own method of parent class A
#ch1.m2() #Error: AttributeError: 'A' object has no attribute 'm2()'
#ch1.m3() #Error: AttributeError: 'A' object has no attribute 'm3()'
````

5. Hybrid Inheritance:
------------------------
- It is a combination of two or more inheritance.
- It may combine hierarchical and multiple inheritance.
- It is more complex than other types of inheritance.

````python
#Inhybrid Inheritance:
class A:
    def m1(self):
        print("m1 (): parent class 1 method")
class B(A):
    def m2(self):
        print("m2 (): child class 1 method")
class C(A):
    def m3(self):
        print("m3 (): child class 2 method")
class D(B,C):
    def m4(self):
        print("m4 (): child class level 2 method")

d = D()
d.m1() #inherited from class A
d.m2() #inherited from class B
d.m3() #inherited from class C
d.m4() #defined in class D

d = B() #object of class B
d.m1() #inherited from class A
d.m2() #own method of class B
#d.m3() #Error: AttributeError: 'B' object has no attribute '
d.m4() #Error: AttributeError: 'B' object has no attribute 'm4()'

d = C() #object of class C
d.m1() #inherited from class A
#d.m2() #Error: AttributeError: 'C' object has no attribute 'm2()'
d.m3() #own method of class C
#d.m4() #Error: AttributeError: 'C' object has no attribute 'm4()'

d = A() #object of class A
d.m1() #own method of class A
#d.m2() #Error: AttributeError: 'A' object has no attribute 'm2()'
#d.m3() #Error: AttributeError: 'A' object has no attribute 'm3()'
#d.m4() #Error: AttributeError: 'A' object has no attribute 'm4()'
````

````python
#Inheritance

class A:
    def m1(self):
        print("class A : m1()")

class B(A):
    def m1(self):
        print("class B : m1()")

s = B()
s.m1() #inherited method from class A #overridden method from class B

s = A()
s.m1() #method of class A 
````

Method Overriding:
- It means child class provides its own version of parent class.
- The same method name is used in both parent and child class.
- Child class method gets priority to execute.

````python
#Inheritance

class A:
    def m1(self):
        print("class A : m1()")

class B(A):
    def m1(self):
        super().m1() #call parent class method
        print("class B : m1()")

s = B()
s.m1() #inherited method from class A #overridden method from class B
````

Is-A Relationship: 
-------------------
- It is a relationship defines inheritance
- One class is a specialized form of another class.
- Child class is a type of parent class.
- Meaning : Single inheritance B(A) : Bclass(child) is a Aclass (parent)
- Ex: Car is a Vehicle, Dog is an Animal, Manerger is an Employee


Has-A Relationship:
--------------------
- It means one class contains object of another class.
- It is achieved using composition and agreegation.
- One object uses another object as a member.
- Meaning : B HAS-A A
- Ex: Car has a Engine, Employee has a Address

````python
class A:
    def m1(self):
        print("Class A : m1()")

class B(A):
    def __init__(self):
        self.s1 = A #B Has-A A

    def m2(self):
        print("class B : m2()")
        self.s1.m1() #B uses A's m1() method

s = B()
s.m1() #Error: AttributeError: 'B' object has no attribute 'm1()' because m1() is not defined in class B, and we are trying to call it on an object of class B.
s.m2()
````
