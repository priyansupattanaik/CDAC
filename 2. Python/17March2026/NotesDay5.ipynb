````python
#@property and @setter
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    @property
    def salary(self):
        return self.__salary
    
    @salary.setter
    def salary(self, amount):
        if amount > 0:
            self.__salary = amount
        else:
            print("Invalid salary amount")

    def display(self):
        print(self.__name)
        print(self.__salary)    

emp = Employee("John Doe", 50000)
emp.display()

print("Current salary:", emp.salary)

emp.salary = 70000
print("Updated Salary:", emp.salary)
````

@property:
----------------------------------------------------------------------------------------
It is a python decorator used to convert a method into a read only attribute.
It allows us to access the method like a variable instead of calling it like a function.
It is commonly used in encapsulation to access private variabes safely.

Key points:
1. It allows method access without parentheses.
2. it works with private variables (__salary).
3. It works together with 
    @property ----> getter
    @property_name.setter ----> setter
    @property_name.deleter ----> deleter

    Syntax:
    @property
    def attribute_name(self):
        return value



````python
class Product:
    def __init__(self, price):
        self.__price = price

    @property
    def price(self):
        return self.__price

p = Product(100)
print(p.price) #access like a variable, but it is actually a method call
````

````python
#AbstractDemo 1

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def display(self):
        pass

class Rectangle(Shape):

        def area(self): #implementation of abstract method
            print("Area of rectangle")

        def display(self): #implementation of abstract method
            print("Area of rectangle hai!")

r = Rectangle()
r.area()
r.display()
````

Abstract Class VS Concrete Class

Abstract Class: 
- A class that cannot be instantiated and is meant to be subclassed. It can contain abstract methods (methods without implementation) that must be implemented by any concrete subclass.
- It serves as a blueprint for other classes and can provide common functionality to its subclasses.
- It is defined using the `abc` module in Python and typically includes the `@abstractmethod` decorator to indicate which methods are abstract.
- Example:
```python
from abc import ABC, abstractmethod
class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass
```

Concrete Class: A class that can be instantiated and provides implementations for all of its methods, including those inherited from abstract classes.
- It can be instantiated to create objects and can be used directly in the program.
- It may inherit from an abstract class and must implement all abstract methods defined in the abstract class
- Example:
```python
class Dog(Animal):
    def make_sound(self):
        return "Woof!"
```



````python
from abc import ABC, abstractmethod

class A(ABC):
    @abstractmethod
    def m1(self):
        pass

    def m2(self):
        print("m2 (): Normal Method")

class B(A):
    def m1(self):
        print("m1 (): implemented and executed")

a = B()
a.m1()
a.m2()
# b = A() #Error: cannot create object of abstract class
# b.m1() #Error: cannot call abstract method
# b.m2() #Error: cannot call method of abstract class
````

````python
#AbstractDemo 3

from abc import ABC, abstractmethod

class A(ABC):
    @abstractmethod
    def m1(self):
        pass

    def m2(self):
        print("m2 (): Normal Method")

class B(A):
    def m1(self):
        print("m1 (): implemented and executed")

    @abstractmethod
    def m3(self):
        pass

a = B()
a.m1()
a.m2()
````

````python
from abc import ABC, abstractmethod

class A(ABC):
    @property
    @abstractmethod
    def m1(self):
        pass

class B(A):
    @property
    def m1(self):
        return "m1 (): implemented as a property in child class"
    
a = B()
print(a.m1)
````

````python
#Abstract Demo 8

from abc import ABC, abstractmethod

class A(ABC):
    @classmethod
    @abstractmethod
    def m1(cls):
        pass

class B(A):
    @classmethod
    def m1(cls):
        print("m1 (): implemented as a class method in child class")

a = B()
a.m1()
````

Abstract Class Key Pointers:

- ABC is used to create abstract base class
- @abstractmethod is used to declare abstract methods
- Abstract class object cannot be created
- Child class must implement all abstract methods
- Abstract class can contain:
    -Normal methods
    -Abstract methods
    -constructors
    -abstract class methods
    -abstract static methods
    -abstract property
