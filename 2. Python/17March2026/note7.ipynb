## 1. Magic Methods (Dunder Methods)

### What are Magic Methods?
- Magic Methods are special methods in Python
- They have double underscore (`__`) before and after the name
- Also called as **dunder methods** (double underscore methods)
- Python calls them automatically in defined situations
- They help objects behave like built-in data types

### Common Magic Methods:
- `__init__` : Called automatically when an object is created
- `__str__` : Called when `print(object)` is used
- `__len__` : Called when `len(object)` is used
- `__add__` : Called when `obj1 + obj2` is used
- `__eq__` : Called when `obj1 == obj2` is used
- `__getitem__` : Called when `obj[index]` is used

### Key Purposes of Magic Methods:
- To initialize objects
- To print objects in a readable form
- To use operators like `+`, `-`, `==`
- To use built-in functions like `len()`, `max()`
- To support indexing like `obj[0]`

````python
# Magic Methods Demonstration
class Test:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __str__(self):
        return f"Employee Name: {self.name}, Salary: {self.salary}"

    def __len__(self):
        return len(self.name)

    def __add__(self, other):
        return self.salary + other.salary


# Creating objects
t1 = Test("Alice", 50000)
t2 = Test("Bob", 60000)

# Testing magic methods
print("__str__ method:", t1)  # Calls __str__
print("__len__ method:", len(t1))  # Calls __len__
print("__add__ method:", t1 + t2)  # Calls __add__
````

## 2. String Formatting Methods

````python
# Different String Formatting Techniques

name = "Amit"
sal = 70000

# Method 1: Using comma separator
print("Method 1 - Comma:", "Name =", name, "& Salary =", sal)

# Method 2: Using + operator
print("Method 2 - + operator:", "Name = " + name)

# Method 3: Using % formatting
print("Method 3 - % formatting:", "Name = %s & Salary = %d" % (name, sal))
price = 123.4567
print("Price with 2 decimals:", "Price = %.2f" % price)

# Method 4: Using format() method
print("Method 4 - format():", "Name = {0}, Salary = {1}".format(name, sal))
print("Named placeholders:", "Name = {n}, Salary = {s}".format(n="Rahul", s=80000))

# Method 5: Using f-string (Python 3.6+)
print("Method 5 - f-string:", f"Name = {name}, Salary = {sal}")
x = 12345.4293923
print(f"Formatted price: Price = {x:.2f}")
print(f"Right aligned (width 10): Price = {sal:10}")
print(f"Left aligned (width 10): Price = {sal:<10}")
print(f"Right aligned (width 10): Price = {sal:>10}")
````

## 3. Inheritance - Hierarchical Inheritance

````python
# Hierarchical Inheritance Demo
# Two classes (Car and Bike) inherit from Vehicle class

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def start(self):
        print(f"Vehicle Start: {self.brand} {self.model}")

    def stop(self):
        print(f"Vehicle Stop: {self.brand} {self.model}")


class Car(Vehicle):
    def __init__(self, brand, model, fuel_type):
        super().__init__(brand, model)  # Inheriting parent class variables
        self.fuel_type = fuel_type

    def display_info(self):
        print("Car Information:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Fuel Type: {self.fuel_type}")


class Bike(Vehicle):
    def __init__(self, brand, model, cc):
        super().__init__(brand, model)  # Inheriting parent class variables
        self.cc = cc

    def display_info(self):
        print("Bike Information:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Capacity: {self.cc}cc")


# Creating objects
c1 = Car("Honda", "Civic", "Gasoline")
b1 = Bike("Yamaha", "R15", 150)

# Testing Car methods
c1.start()
c1.display_info()
c1.stop()

print("\n" + "-"*30 + "\n")

# Testing Bike methods
b1.start()
b1.display_info()
b1.stop()
````

## 4. HAS-A Relationship Types

### 4.1 Association
- A general relationship where one object is connected to another
- Example: A teacher teaches students
- Both classes can exist independently
- Most general form of relationship

````python
# Association Example

class Student:
    def __init__(self, name):
        self.name = name

class Teacher:
    def __init__(self, name):
        self.name = name

    def teach(self, student):
        print(f"{self.name} teaches {student.name}")


# Creating objects
s1 = Student("Alice")
t1 = Teacher("Mr. Smith")

# Association is created at runtime
t1.teach(s1)
````

### 4.2 Aggregation (Weak HAS-A)
- A weak HAS-A relationship
- One object uses or contains another object, but the contained object can exist independently
- Example: A department has employee objects
- Follows one-to-many form

````python
# Aggregation Example

class Employee:
    def __init__(self, name):
        self.name = name

class Department:
    def __init__(self, dname):
        self.dname = dname
        self.employees = []  # Empty list to store employee objects

    def add_emp(self, employee):  # employee is an object of Employee class
        self.employees.append(employee)

    def display(self):
        print(f"Department: {self.dname}")
        for emp in self.employees:
            print(f"  - {emp.name}")


# Creating objects
e1 = Employee("Amit")
e2 = Employee("Sara")

d1 = Department("Education")
d1.add_emp(e1)
d1.add_emp(e2)

d1.display()

# Employees exist independently even if department is deleted
print(f"\nEmployees still exist: {e1.name}, {e2.name}")
````

### 4.3 Composition (Strong HAS-A)
- A strong HAS-A relationship
- The contained object is strongly owned by the container
- If the container is destroyed, the contained object is also destroyed
- Example: A house has rooms

````python
# Composition Example

class Room:
    def __init__(self, room_type):
        self.room_type = room_type

class House:
    def __init__(self):
        # Rooms are created inside House constructor
        # Rooms are strongly owned by House
        self.r1 = Room("Kitchen")
        self.r2 = Room("Balcony")

    def display(self):
        print(self.r1.room_type)
        print(self.r2.room_type)


# Creating House object (rooms are created automatically)
h1 = House()
h1.display()

# If h1 is deleted, rooms are also destroyed
print("\nNote: Rooms are destroyed when House is destroyed")
````

### 4.4 HASA (Composition with Engine example)

````python
# HASA Demo - Composition

class Engine:
    def __init__(self, engine_no, power):
        self.engine_no = engine_no
        self.power = power

    def start(self):
        print(f"Engine {self.engine_no} with {self.power} HP started.")

class Car:
    def __init__(self, brand, model, engine):
        self.engine = engine  # Car HAS-A Engine (composition)
        self.brand = brand
        self.model = model

    def start1(self):
        print(f"Brand name = {self.brand} & Model name = {self.model}")
        print(f"Having engine no = {self.engine.engine_no}")
        self.engine.start()

    def display_info(self):
        print("Car Information:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Engine No: {self.engine.engine_no}")
        print(f"Power: {self.engine.power} HP")


# Creating objects
e1 = Engine("ENG123", 150)
c1 = Car("Toyota", "Corolla", e1)

c1.start1()
print()
c1.display_info()
````

### Comparison Table: Association vs Aggregation vs Composition

| Relationship Type | Description | Example |
|-------------------|-------------|---------|
| **Association** | A general relationship where one object is connected to another. Both classes can exist independently. | A teacher teaches students. |
| **Aggregation** | A weak HAS-A relationship where one object uses or contains another object, but the contained object can exist independently. | A department has employee objects. |
| **Composition** | A strong HAS-A relationship where the contained object is strongly owned by the container. If the container is destroyed, the contained object is also destroyed. | A house has rooms. |

## 5. Polymorphism

### Definition:
- **Poly** = many, **morphism** = forms
- Polymorphism is the ability of the same operation to behave differently for different objects

### Need of Polymorphism:
- Improves code reusability
- Allows same interface for different objects
- Supports extensibility
- Important concept of OOP

### Types of Polymorphism:

#### 1. Compile Time Polymorphism (Static Polymorphism / Early Binding)
   - Method Overloading
   - Operator Overloading
   - Occurs at compile time

#### 2. Run Time Polymorphism (Dynamic Polymorphism / Late Binding)
   - Method Overriding
   - Occurs at run time

````python
# 5.1 Method Overriding (Runtime Polymorphism)

class A:
    def m1(self):
        print("Class A : m1()")

class B(A):
    def m1(self):  # Overriding parent's m1 method
        print("Class B : m1() : child class")


b1 = B()
b1.m1()  # Calls child class method (overridden method)

# If B didn't have its own m1, it would call A's m1
````

````python
# 5.2 Operator Overloading (Compile Time Polymorphism)

print("Integer addition:", 10 + 20)
print("String concatenation:", "Hello " + "World")
print("List concatenation:", [1, 2] + [3, 4])

# The + operator behaves differently based on operand types
````

````python
# 5.3 Method Overloading Style using Default Arguments

class A:
    def m1(self, a=None, b=None):
        if a is not None and b is not None:
            print(a + b)
        elif a is not None:
            print(a)
        else:
            print("Arey bhai koi argument nahi hai!")


a1 = A()
print("No argument:", end=" ")
a1.m1()  # No argument
print("One argument:", end=" ")
a1.m1(10)  # One argument
print("Two arguments:", end=" ")
a1.m1(10, 20)  # Two arguments
````

````python
# 5.4 Method Overloading Style using *args

class A:
    def m1(self, *args):
        print("Arguments are:", args)


a1 = A()
a1.m1()
a1.m1(10)
a1.m1(10, 20)
a1.m1(10, 20, 30, 40)
````

````python
# 5.5 Duck Typing Polymorphism

class A:
    def m1(self):
        print("Class A : m1()")

class B:
    def m1(self):
        print("Class B : m1()")

def call(obj):
    obj.m1()  # Works as long as obj has m1() method


call(A())  # Calls A's m1
call(B())  # Calls B's m1
````

````python
# 5.6 Polymorphism with Built-in Functions

print("len() with string:", len("Sona math!!!!!!"))
print("len() with list:", len([1, 2, 3, 4, 5]))
print("len() with tuple:", len((1, 2, 3, 4, 5)))

# len() behaves differently based on input type
````

````python
# 5.7 Polymorphism with Inheritance Hierarchy

class A:
    def m1(self):
        print("Class A : m1()")

class B(A):
    def m1(self):
        print("Class B : m1()")

class C(A):
    def m1(self):
        print("Class C : m1()")


obj = [B(), C()]
for o1 in obj:
    o1.m1()  # Calls appropriate method based on object type
````

## 6. Exception Handling

### Error vs Exception

| Aspect | Error | Exception |
|--------|-------|-----------|
| **Definition** | An error is a problem in a program that causes it to behave unexpectedly or stop. | An exception is a type of runtime error that occurs during program execution. |
| **Occurrence** | Errors can happen before the program runs or while it executes. | Exceptions occur during program execution. |
| **Handling** | Errors indicate something is wrong in the code or logic and are typically not handled. | Exceptions can be handled using try-except blocks to prevent the program from stopping. |

### Types of Errors:

#### 1. Syntax Error:
- Occurs when Python rules are violated
- Example:
```python
if 10 > 5  # Missing colon
    print("Hello")
```

#### 2. Indentation Error:
- Wrong indentation causes IndentationError
- Example:
```python
def hi():
print("Hi")  # Missing indentation
```

#### 3. Exceptions (Runtime Errors):
- Occur after program starts execution

### Common Exceptions:
- **ZeroDivisionError**: division by zero
- **ValueError**: invalid value
- **TypeError**: invalid type operation
- **IndexError**: list index out of range
- **KeyError**: dictionary key not found
- **FileNotFoundError**: file doesn't exist
- **NameError**: variable not found
- **AttributeError**: attribute/method not found

### Exception Hierarchy:
```
BaseException
    ├── SystemExit
    ├── KeyboardInterrupt
    └── Exception
         ├── ArithmeticError
         │    ├── ZeroDivisionError
         │    ├── OverflowError
         │    └── FloatingPointError
         └── LookupError
              ├── IndexError
              └── KeyError
```

### Main Keywords:
- **try**: handle risky code
- **except**: handles specific exception
- **else**: runs if no exception occurs
- **finally**: always executes
- **raise**: used to explicitly raise exception

### Syntax:
```python
try:
    # Risky code
except SomeException:
    # Handle exception
else:
    # Runs if no exception occurs
finally:
    # Always executes
```

````python
# Exception Demo 1 - Basic try-except

n = 10
try:
    print("start...........")
    res = n / 0  # This will raise ZeroDivisionError
    print("stop...........")  # This line won't execute
except ZeroDivisionError:
    print("Cannot divide by zero!")

print("end...........")
````

````python
# Exception Demo 2a - try-except-else-finally

try:
    print("start...........")
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    result = a / b
    print(f"Result: {result}")
    print("stop...........")
except ZeroDivisionError:
    print("Cannot divide by zero!")
else:
    print("No Exception occurred")
finally:
    print("Finally block - Always executed")

print("end...........")
````

````python
# Exception Demo 2b - try-except-finally (without else)

try:
    print("start...........")
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    result = a / b
    print(f"Result: {result}")
    print("stop...........")
except ZeroDivisionError:
    print("Cannot divide by zero!")
finally:
    print("Finally block - Always executed")

print("end...........")
````
