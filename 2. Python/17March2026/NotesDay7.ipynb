**Magic Methods:**

- Magic Methods are special methods.
- They have double underscore before and after the name.
- They are also called as dunder method.
- Python programs call them automatically, in defined situations.
- They help objects behave like built-in data types.

- Ex:
    __init__ :called automatically when an object is created.
    __str__ : called when we print(object) is used
    __len__ : called when len(object) is used
    __add__ : called when obj1 + obj2 is used
    __eq__ : called when obj1 == obj2 is used
    __getitem__ : called when obj[index] is used


**Key Pointer to use Magic Methods:**
- To initialize objects
- To print objects in a readable form
- To use operator, like +, -, ==
- To use built-in functions, like len(), max()
- To support indexing like obj[0]

- Ex:
```python
def __str__(self):
    return "Result= " + self.name

````python
#MagicDemo - 1

class Test:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __str__(self):
        return f"Employee Name: {self.name}, Salary: {self.salary}"

def __len__(self):
    return len(self.name)

def __add__(self, other):
    return self.salary + other.salary


t1 = Test("Alice", 50000)
t2 = Test("Bob", 60000)
print(t1)  # Employee Name: Alice, Salary: 50000
print(len(t1))  # 5
print(t1 + t2)  # 110000
````

````python
#ISADemo
 #their are two class vehicles and car and define two methods start() and stop() in vehicle class

#hirarcial inheritance demo

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def start(self):
        print(f"Vehicle Start: {self.brand} {self.model}")

    def stop(self):
        print(f"Vehicle Stop: {self.brand} {self.model}")

class Car(Vehicle):
    def __init__(self, brand, model, fuel_type):
        super().__init__(brand, model) #Inheriting parent class variables
        self.fuel_type = fuel_type

    def display_info(self):
        print("Car Information:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Fuel Type: {self.fuel_type}")

class Bike(Vehicle):
    def __init__(self, brand, model, cc):
        super().__init__(brand, model) #Inheriting parent class variables
        self.cc = cc

    def display_info(self):
        print("Bike Information:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Capacity: {self.cc}")


c1 = Car("Honda", "Civic", "Gasoline")
b1 = Bike("Yamaha", "R15", 150)

c1.start()
c1.display_info()
c1.stop()

print()

b1.start()
b1.display_info()
b1.stop()
````

**Association**
- A general relationship where one object is connected to another.
- Ex: A teacher teaches students
- Both Classes can exist independently
- Most general form of relationship

````python
#Association Example

class Student:
    def __init__(self, name):
        self.name = name

class Teacher:
    def __init__(self, name):
        self.name = name

    def teach(self, student):
        print(f"{self.name} teaches {student.name}")

s1 = Student("Alice")
t1 = Teacher("Mr. Smith")

t1.teach(s1) #association is created at s1.
````

**Aggregation**
- A weak HAS-A relationship is represented by aggregation.
- One object uses or contains another object, but the contained object can exist independently.
- Ex: A department has employee objects
- It follows one to many form

````python
#Aggregation Example
class Employee: 
    def __init__(self, name):
        self.name = name

class Department:
    def __init__(self, dname):
        self.dname = dname
        self.employees = [] # #employees is a list to hold employee objects #created empty list to store employee objects

    def add_emp(self, employee): #employee is an object of employee class
        self.employees.append(employee) #employee class object is passes to add_emp method and added to employees list

    def display(self):
        print(f"Department: {self.dname}")
        for emp in self.employees: #this will run till all the objects of employee class are printed
            print(emp.name) 

e1 = Employee("Amit")
e2 = Employee("Sara")

d1 = Department("Education")
d1.add_emp(e1)
d1.add_emp(e2)

d1.display()
    
````

**Compostion**
- A strong HAS-A relationship is represented by composition.
- The contained object is strongly owned by the container and its container.
- Ex: A house has rooms, if the house is destroyed, the rooms are also destroyed.
- It follows one to many form

````python
#Composition Example

class Room:
    def __init__(self, room_type):
        self.room_type = room_type

class House:                     
    def __init__(self):           #house class constructor is creating room class objects
        self.r1 = Room("Kitchen") #creating room class object in house class constructor
        self.r2 = Room("Balcony") #creating room class object in house class constructor

    def display(self): #house class method to access room class variables
        print(self.r1.room_type) #accessing room class variable in house class method 
        print(self.r2.room_type) #accessing room class variable in house class method

h1 = House() #creating house class object
h1.display() #accessing house class method which is accessing room class variables
````

Difference bewteen Association, Aggregation and Composition:

| Relationship Type | Description | Example |
|-------------------|-------------|---------|
| Association | A general relationship where one object is connected to another. Both classes can exist independently. | A teacher teaches students. |
| Aggregation | A weak HAS-A relationship where one object uses or contains another object, but the contained object can exist independently. | A department has employee objects. |
| Composition | A strong HAS-A relationship where the contained object is strongly owned by the container and its container. If the container is destroyed, the contained object is also destroyed. | A house has rooms. If the house is destroyed, the rooms are also destroyed. |

**4. Polymorphism:**
- Polymorphism is the ability of same operation to behave differently for different objects.
- Poly = many, morphism = forms
- Need of Polymorphism:
    - It improves code reusability.
    - It allows same interface for different objects.
    - It supports extensibility
    - It is one of the important concept of OOP.

Types of Polymorphism:
1. Compile Time Polymorphism (Method Overloading)
    - Also called as Static Polymorphism. (Early Binding)
    - Occurs at compile time
    - Method Overloading, Operator Overloading

2. Run Time Polymorphism (Method Overriding)
    - Also called as Dynamic Polymorphism. (Late Binding)
    - Occurs at run time
    - Method Overriding

````python
#Ploymorphism Example

class A:
    def m1(self, a= None, b= None):
        if a is not None and b is not None:
            print(a + b)
        elif a is not None:
            print(a) 
        else:
            print("Arey bhai koi argument nahi hai!")

a1 = A()
a1.m1() #No argument
a1.m1(10) #One argument
a1.m1(10, 20) #Two arguments
````

````python
#Polymorphism Demo 2

class A:
    def m1(self):
        print("Class A : m1()")

class B(A):
    def m1(self):
        print("Class B : m1() : child class")

b1 = B()
b1.m1() #This will call the m1 method of class B, demonstrating polymorphism. 
        #If class B did not have its own m1 method, it would have called the m1 method of class A.
````

5. Polymorphsim: through duck typing

````python
class A:
    def m1(self):
        print("Class A : m1()")

class B:
    def m1(self):
        print("Class B : m1()")

def call(obj):
    obj.m1()

call(A()) #This will call the m1 method of class A, demonstrating polymorphism.
call(B()) #This will call the m1 method of class B, demonstrating polymorphism.
````

Exception Handling:
1. Error: 
    -   An error is a probleme in a program that causes the program to behave unexpectedly or stop.
    -   Errors can happen:
        - before the program run
        - while the program executes
    - Error indicate something is wrong in the code or logic.

2. Exception:
    - An exception is a type of runtime error.
    - It occurs during program execution
    - It exeception is not handled, it stops the program.

COMPARE ERROR AND EXCEPTION:
| Aspect | Error | Exception |
|--------|-------|-----------|
| Definition | An error is a problem in a program that causes it to behave unexpectedly or stop. | An exception is a type of runtime error that occurs during program execution. |
| Occurrence | Errors can happen before the program runs or while it executes. | Exceptions occur during program execution. |
| Handling | Errors indicate something is wrong in the code or logic and are typically not handled. | Exceptions can be handled using try-except blocks to prevent the program from stopping. |
