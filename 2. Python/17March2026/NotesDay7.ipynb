**Magic Methods:**

- Magic Methods are special methods.
- They have double underscore before and after the name.
- They are also called as dunder method.
- Python programs call them automatically, in defined situations.
- They help objects behave like built-in data types.

- Ex:
    __init__ :called automatically when an object is created.
    __str__ : called when we print(object) is used
    __len__ : called when len(object) is used
    __add__ : called when obj1 + obj2 is used
    __eq__ : called when obj1 == obj2 is used
    __getitem__ : called when obj[index] is used


**Key Pointer to use Magic Methods:**
- To initialize objects
- To print objects in a readable form
- To use operator, like +, -, ==
- To use built-in functions, like len(), max()
- To support indexing like obj[0]

````python
#MagicDemo - 1

class Test:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __str__(self):
        return f"Employee Name: {self.name}, Salary: {self.salary}"

def __len__(self):
    return len(self.name)

def __add__(self, other):
    return self.salary + other.salary


t1 = Test("Alice", 50000)
t2 = Test("Bob", 60000)
print(t1)  # Employee Name: Alice, Salary: 50000
print(len(t1))  # 5
print(t1 + t2)  # 110000
````

````python
#ISADemo
 #their are two class vehicles and car and define two methods start() and stop() in vehicle class

#hirarcial inheritance demo

class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def start(self):
        print(f"Vehicle Start: {self.brand} {self.model}")

    def stop(self):
        print(f"Vehicle Stop: {self.brand} {self.model}")

class Car(Vehicle):
    def __init__(self, brand, model, fuel_type):
        super().__init__(brand, model) #Inheriting parent class variables
        self.fuel_type = fuel_type

    def display_info(self):
        print("Car Information:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Fuel Type: {self.fuel_type}")

class Bike(Vehicle):
    def __init__(self, brand, model, cc):
        super().__init__(brand, model) #Inheriting parent class variables
        self.cc = cc

    def display_info(self):
        print("Bike Information:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Capacity: {self.cc}")


c1 = Car("Honda", "Civic", "Gasoline")
b1 = Bike("Yamaha", "R15", 150)

c1.start()
c1.display_info()
c1.stop()

print()

b1.start()
b1.display_info()
b1.stop()
````

````python
#HASA demo

class Engine:
    def __init__(self,engine_no, power):
        self.engine_no = engine_no
        self.power = power

    def start(self):
        print(f"Engine {self.engine_no} with {self.power} HP started.")

class Car:
    def __init__(self, brand, model, engine):
        self.engine = engine
        self.brand = brand
        self.model = model

    def start1(self):
        print(f"Brand name = {self.brand} & Model name = {self.model} and having engine  = {self.engine.engine_no}")
        self.engine.start()

    def display_info(self):
        print("Car Information:")
        print(f"Brand: {self.brand}")
        print(f"Model: {self.model}")
        print(f"Engine No: {self.engine}") #Engineclass object address
        print(f"Power: {self.engine.power} HP") #In car class we are accessing engine

e1 = Engine("ENG123", 150) 
c1 = Car("Toyota", "Corolla", e1) #e1 = connection of engine class
c1.start1()
c1.display_info()
````

**Association**
- A general relationship where one object is connected to another.
- Ex: A teacher teaches students
- Both Classes can exist independently
- Most general form of relationship

````python
#Association Example

class Student:
    def __init__(self, name):
        self.name = name

class Teacher:
    def __init__(self, name):
        self.name = name

    def teach(self, student):
        print(f"{self.name} teaches {student.name}")

s1 = Student("Alice")
t1 = Teacher("Mr. Smith")

t1.teach(s1) #association is created at s1.
````

**Aggregation**
- A weak HAS-A relationship is represented by aggregation.
- One object uses or contains another object, but the contained object can exist independently.
- 

````python
#Aggregation Example
class Employee: 
    def __init__(self, name):
        self.name = name

class Department:
    def __init__(self, dname):
        self.dname = dname
        self.employees = [] # #employees is a list to hold employee objects #created empty list to store employee objects

    def add_emp(self, employee): #employee is an object of employee class
        self.employees.append(employee) #employee class object is passes to add_emp method and added to employees list

    def display(self):
        print(f"Department: {self.dname}")
        for emp in self.employees: #this will run till all the objects of employee class are printed
            print(emp.name)

e1 = Employee("Amit")
e2 = Employee("Sara")

d1 = Department("Education")
d1.add_emp(e1)
d1.add_emp(e2)

d1.display()
    
````
