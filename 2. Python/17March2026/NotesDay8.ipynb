Exception Chaining:
- When one exception is raised while handling another exception.
- It helps to preserve the original error context.
- It can be implemented using the keyword:

    raise..... from.....

- It is used to show the root cause of the error.
- It also improves the debugging clarity which helps in tracking error flow.

Syntax:
```python
raise NewException("message") from OriginalException
```

````python
#Exception Demo 17

def divide(a, b):
    try:
        return a/b
    except ZeroDivisionError as e:
        raise ValueError("Cannot divide by zero") from e
    
divide(10, 0)
````

````python
#Exception Demo 17 #without chaining

def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero")
    return a / b
    
try:
        divide(10, 0)
except ValueError as e:
        print(e)
````

````python
#Exception Demo 18

class InsufficientBalanceError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientBalanceError("Not sufficient balance.......")
    
    return balance - amount

try:
    balance = withdraw(10000, 1500)

except InsufficientBalanceError as e:
    print("Custom Exception is defined..........", e)
````

**File Handling:**
------------------

- It is the process of creating, opening, reading, writing and closing files.
- Purpose: store the data permanently.
- RAM --> Disk
- Need :
    - Ex: student data, employee records, log, reports, configuration files,... data is required to store.
    - Different types of fils: txt, csv, json, sql,...

Opening a file:
```python
 file_object = open("fileName", "mode")
```

Mode of file:
1. Read : r
    - Opens file for reading.
    - File must be existing.
    - Default mode is read mode.
- Ex:
    ```python
    f = open("abc.txt", "r")
    print(f.read())
    f.close()
    ```

2. Write : w
    - Open file for witing.
    - It creates file if it doesnot exist.
    - If file already exists, old content is erased.
- Ex:
    ```python
    f = open("abc.txt", "w")
    f.write("Hello World")
    f.close()
    ```
3. Append : a
    - Open file for appending
    - Creates file if it does not exist.
    - New data is added at the end of the file
- Ex:
    ```python
    f = open("abc.txt", "a")
    f.write("Hello CDAC")
    f.close()
    ```
4. Exclusive create mode : x
    - Create a new file
    - If file already exist ---> Error
- Ex:
    ```python
    f = open("abc.txt", "x")
    f.write("Hello CDAC")
    f.close()
    ```
5. Binary Mode: Non-text file
- images, pdf, audio, video etc
- 'b' is added in the mode
- r: rb, w: wb, a:ab

6. Combine File Mode:
- Read : r+
     - Read and write
     - File must exist
     - Pointer starts at the beginning of 1st character

- Write : w+
    - Read and write
    - Overrites existing file
    - Creates file if it does not exist

- Append : a+
    - Read and Appen
    - Creates file if it does not exist
    - Writing always happens at the end of the file

1. tell() : return current postion
2. seek() : moves pointer to a specific position

````python
import os

with open ("abcde.txt", "r") as f:
   print(f.tell()) #starting position = 0
   print(f.read(5))
   print(f.tell()) # position after reading 5 character
   f.seek(0) # move pointer to the beginning of the file
   print(f.read())

   if os.path.exists("abcde.txt"):
       print("File exists")
   else:
       print("File does not exist")
   print(f.name)
   print(f.mode)
   print(f.closed)
````
