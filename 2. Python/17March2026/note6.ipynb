# Part 1: Encapsulation

Encapsulation is the bundling of data (attributes) and methods that operate on that data within a single unit (class). It also restricts direct access to some of an object's components.

## Basic Example Without Encapsulation

````python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display(self):
        print("Name:", self.name)
        print("Salary:", self.salary)
````

## Encapsulation with Private Variables

In Python, we use double underscore `__` to make attributes private.

````python
class Employee:
    def __init__(self, name, salary):
        self.__name = name        # Private variable
        self.__salary = salary    # Private variable

    def display(self):
        print(self.__name)
        print(self.__salary)

emp = Employee("John Doe", 50000)
emp.display()
# print(emp.__name)  # This will raise an AttributeError - cannot access private variables directly
````

## Using Getter and Setter Methods

Getter and setter methods provide controlled access to private variables.

````python
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    # Getter method : read data
    def get_salary(self):
        return self.__salary
    
    # Setter method : write data with validation
    def set_salary(self, amount):
        if amount > 0:
            self.__salary = amount
        else:
            print("Invalid salary amount")

    def display(self):
        print(self.__name)
        print(self.__salary)

emp = Employee("John Doe", 50000)
emp.display()
emp.set_salary(70000)
print("Updated Salary:", emp.get_salary())
````

## Pythonic Approach: @property and @setter Decorators

The `@property` decorator allows us to define methods that can be accessed like attributes.

````python
class Employee:
    def __init__(self, name, salary):
        self.__name = name
        self.__salary = salary

    @property
    def salary(self):
        return self.__salary
    
    @salary.setter
    def salary(self, amount):
        if amount > 0:
            self.__salary = amount
        else:
            print("Invalid salary amount")

    def display(self):
        print(self.__name)
        print(self.__salary)    

emp = Employee("John Doe", 50000)
emp.display()
print("Current salary:", emp.salary)  # Accessed like a property, not a method
emp.salary = 70000  # Using setter like attribute assignment
print("Updated Salary:", emp.salary)
````

### Understanding @property Decorator

**Key Points:**
1. Converts a method into a **read-only attribute**
2. Allows method access **without parentheses**
3. Commonly used to access private variables safely
4. Works with three decorators:
   - `@property` → Getter
   - `@<property_name>.setter` → Setter
   - `@<property_name>.deleter` → Deleter

**Syntax:**
```python
@property
def attribute_name(self):
    return value
```

````python
class Product:
    def __init__(self, price):
        self.__price = price

    @property
    def price(self):
        return self.__price

p = Product(100)
print(p.price)  # Access like a variable, but it's actually a method call
````

````python
class Employee:
    def __init__(self, name, salary, password):
        self.__name = name
        self.__salary = salary
        self.__password = password

    def display(self):
        print(self.__name)
        print(self.__salary)

emp = Employee("John Doe", 50000, "secret123")
emp.display()
````

# Part 2: Abstraction

Abstraction hides complex implementation details and shows only essential features. In Python, we achieve abstraction using Abstract Base Classes (ABC).

````python
from abc import ABC, abstractmethod

class A(ABC):  # Abstract class (inherits from ABC)
    @abstractmethod
    def m1(self):  # Abstract method
        pass 

class B(A):  # Concrete class
    def m1(self):
        print("m1(): implemented in child class")

a = B()
a.m1()
````

````python
from abc import ABC, abstractmethod

class A(ABC):  # Abstract class
    @abstractmethod
    def m1(self):
        pass

class B(A):  # Concrete class
    def m1(self):
        print("m1(): implemented in child class")

# a = A()  # Error: Cannot instantiate abstract class
a = B()
a.m1()
````

## Abstract Class with Multiple Abstract Methods

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def display(self):
        pass

class Rectangle(Shape):
    def area(self):  # Implementation of abstract method
        print("Area of rectangle")

    def display(self):  # Implementation of abstract method
        print("Displaying rectangle!")

r = Rectangle()
r.area()
r.display()
````

## Abstract Class with Normal Methods

````python
from abc import ABC, abstractmethod

class A(ABC):
    @abstractmethod
    def m1(self):
        pass

    def m2(self):  # Normal method in abstract class
        print("m2(): Normal Method")

class B(A):
    def m1(self):
        print("m1(): implemented and executed")

a = B()
a.m1()
a.m2()
````

## Abstract Class with Abstract Properties

````python
from abc import ABC, abstractmethod

class A(ABC):
    @property
    @abstractmethod
    def m1(self):
        pass

class B(A):
    @property
    def m1(self):
        return "m1(): implemented as a property in child class"
    
a = B()
print(a.m1)
````

## Abstract Class with Abstract Class Methods

````python
from abc import ABC, abstractmethod

class A(ABC):
    @classmethod
    @abstractmethod
    def m1(cls):
        pass

class B(A):
    @classmethod
    def m1(cls):
        print("m1(): implemented as a class method in child class")

a = B()
a.m1()
````

## Abstract Class vs Concrete Class

| Abstract Class | Concrete Class |
|----------------|-----------------|
| Cannot be instantiated | Can be instantiated |
| Can contain abstract methods | All methods have implementation |
| Serves as a blueprint | Can be used directly |
| Must be inherited | May inherit from abstract classes |

**Abstract Class Key Pointers:**
- `ABC` is used to create abstract base class
- `@abstractmethod` declares abstract methods
- Abstract class objects cannot be created
- Child classes must implement all abstract methods
- Abstract classes can contain:
  - Normal methods
  - Abstract methods
  - Constructors
  - Abstract class methods
  - Abstract static methods
  - Abstract properties

## Practical Example: Vehicle Abstract Class

````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

class Motorbike(Vehicle):
    def start(self):
        print("Motorbike started")

    def stop(self):
        print("Motorbike stopped")

c = Motorbike()
c.start()
c.stop()
````

# Part 3: Inheritance

Inheritance is an OOP concept where one class (child class) acquires the properties and behaviors (attributes and methods) of another class (parent class).

## Inheritance Terminology

- **Parent Class / Base Class / Super Class**: The class being inherited from
- **Child Class / Derived Class / Sub Class**: The class that inherits

**Real World Example:**
- Parent class: Vehicle
- Child classes: Car, Bike, Truck

**Syntax:**
```python
class ParentClass:
    # parent class code

class ChildClass(ParentClass):
    # child class code
```

**Advantages:**
- Code reuse
- Reduce code duplication
- Easy maintenance
- Logical program structure
- Supports hierarchical relationships

## Types of Inheritance

1. **Single Inheritance**
2. **Multiple Inheritance**
3. **Multilevel Inheritance**
4. **Hierarchical Inheritance**
5. **Hybrid Inheritance**

### 1. Single Inheritance

A child class inherits from only one parent class.

````python
class A:
    def m1(self):
        print("m1(): parent class method")

class B(A):
    def m2(self):
        print("m2(): child class method")

print("--- Child class object ---")
s = B()  # Object of child class
s.m1()   # Inherited method from parent class
s.m2()   # Child class method

print("\n--- Parent class object ---")
s = A()  # Object of parent class
s.m1()   # Parent class method
# s.m2()  # Error: 'A' object has no attribute 'm2'
````

### 2. Multiple Inheritance

A child class inherits from more than one parent class.

````python
class A:
    def m1(self):
        print("m1(): parent class 1")

class B:
    def m2(self):
        print("m2(): parent class 2")

class C(A, B):
    def m3(self):
        print("m3(): child class method")

print("--- Child class object ---")
c = C()
c.m1()
c.m2()
c.m3()

print("\n--- Parent class 1 object ---")
c = A()
c.m1()
# c.m2()  # Error
# c.m3()  # Error

print("\n--- Parent class 2 object ---")
c = B()
# c.m1()  # Error
c.m2()
# c.m3()  # Error
````

### 3. Multilevel Inheritance

Inheritance occurs in multiple levels - a class inherits from another derived class.

````python
class A:
    def m1(self):
        print("m1(): parent class 1")

class B(A):
    def m2(self):
        print("m2(): parent class 2")

class C(B):
    def m3(self):
        print("m3(): child class method")

print("--- Child class C object ---")
c1 = C()
c1.m1()  # Inherited from A
c1.m2()  # Inherited from B
c1.m3()  # Own method

print("\n--- Parent class B object ---")
c1 = B()
c1.m1()  # Inherited from A
c1.m2()  # Own method of B
# c1.m3()  # Error

print("\n--- Parent class A object ---")
c1 = A()
c1.m1()  # Own method of A
# c1.m2()  # Error
# c1.m3()  # Error
````

### 4. Hierarchical Inheritance

Multiple child classes inherit from one parent class.

````python
class A:
    def m1(self):
        print("m1(): parent class method")

class B(A):
    def m2(self):
        print("m2(): child class 1 method")

class C(A):
    def m3(self):
        print("m3(): child class 2 method")

print("--- Child class B object ---")
ch1 = B()
ch1.m1()  # Inherited from A
ch1.m2()  # Own method
# ch1.m3()  # Error

print("\n--- Child class C object ---")
ch1 = C()
ch1.m1()  # Inherited from A
# ch1.m2()  # Error
ch1.m3()  # Own method

print("\n--- Parent class A object ---")
ch1 = A()
ch1.m1()  # Own method
# ch1.m2()  # Error
# ch1.m3()  # Error
````

### 5. Hybrid Inheritance

Combination of two or more types of inheritance.

````python
class A:
    def m1(self):
        print("m1(): parent class 1 method")

class B(A):
    def m2(self):
        print("m2(): child class 1 method")

class C(A):
    def m3(self):
        print("m3(): child class 2 method")

class D(B, C):
    def m4(self):
        print("m4(): child class level 2 method")

print("--- Class D object (hybrid) ---")
d = D()
d.m1()  # From A
d.m2()  # From B
d.m3()  # From C
d.m4()  # From D

print("\n--- Class B object ---")
d = B()
d.m1()  # From A
d.m2()  # From B
# d.m3()  # Error
# d.m4()  # Error

print("\n--- Class C object ---")
d = C()
d.m1()  # From A
# d.m2()  # Error
d.m3()  # From C
# d.m4()  # Error

print("\n--- Class A object ---")
d = A()
d.m1()  # From A
# d.m2()  # Error
# d.m3()  # Error
# d.m4()  # Error
````

**MRO - Method Representation object is the order in which Python looks for a method in a hierarchy of classes. It follows the C3 linearization algorithm to determine the method resolution order.**

## Method Overriding

Method overriding occurs when a child class provides its own implementation of a method that is already defined in the parent class.

````python
class A:
    def m1(self):
        print("class A : m1()")

class B(A):
    def m1(self):  # Overriding parent method
        print("class B : m1()")

s = B()
s.m1()  # Calls overridden method from class B

s = A()
s.m1()  # Calls original method from class A
````

## Using super() to Call Parent Class Methods

`super()` is used to call parent class methods from the child class. It helps avoid code duplication and allows extending parent behavior.

````python
class A:
    def m1(self):
        print("class A : m1()")

class B(A):
    def m1(self):
        super().m1()  # Call parent class method
        print("class B : m1()")

s = B()
s.m1()  # Calls both parent and child methods
````

## Constructor Chaining with super()

````python
class A:
    def __init__(self):
        print("Constructor : A")

class B(A):
    def __init__(self):
        super().__init__()
        print("Constructor : B")

    def m1(self):
        print("m1(): method of class B")

s = B()
s.m1()
````

````python
class A:
    def m1(self):
        print("Class A : m1()")

class B(A):
    def m2(self):
        print("class B : m2()")

s = B()
s.m1()
s.m2()
````

## Is-A vs Has-A Relationship

### Is-A Relationship
- Defines inheritance
- One class is a specialized form of another class
- Child class **is a** type of parent class
- Syntax: `class B(A):` → B **is a** A
- Examples:
  - Car **is a** Vehicle
  - Dog **is an** Animal
  - Manager **is an** Employee

### Has-A Relationship
- One class contains object of another class
- Achieved using composition and aggregation
- One object **uses** another object as a member
- Syntax: B **has a** A
- Examples:
  - Car **has an** Engine
  - Employee **has an** Address

````python
# Has-A Relationship Example
class A:
    def m1(self):
        print("Class A : m1()")

class B(A):
    def __init__(self):
        self.s1 = A  # B Has-A A (composition)

    def m2(self):
        print("class B : m2()")
        self.s1.m1()  # B uses A's m1() method

s = B()
s.m2()
````

## Complex Inheritance Example with Multiple Classes

````python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
        print("Constructor : Employee")

class Manager(Employee):
    def __init__(self, name, salary):
        super().__init__(name, salary)
        print("Constructor : Manager")

    def m1(self):
        print("m1(): Manager method")

class Admin(Manager, Employee):
    def __init__(self, name, salary):
        super().__init__(name, salary)
        print("Constructor : Admin")

    def m2(self):
        print("m2(): Admin method")

print("--- Creating objects ---")
e = Employee("John", 50000)
print()
m = Manager("Alice", 70000)
print()
a = Admin("Bob", 60000)
````
