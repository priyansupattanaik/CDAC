# Python Dictionaries (key:value pairs)

## Basic Dictionary Creation

```python
# Creating a dictionary
student = {"name" : "Amit", "addr" : "mumbai", "course" : "AI"}
print(student)

# Accessing values using keys
print(student["name"])
```

## Valid Dictionary Keys
- Allowed: int, float, str, tuple, frozenset
- Not Allowed: list, set, dictionary (mutable types)

```python
# This will raise TypeError: unhashable type: 'list'
# d = {[1,2]: "a"}  # Uncomment to see error

# Valid key types
valid_dict = {
    1: "integer key",
    3.14: "float key",
    "name": "string key",
    (1,2): "tuple key"
}
print(valid_dict)
```

## Dictionary Constructor - dict()

```python
# Creating dictionary from list of tuples
d = dict([("a", 1), ("b", 2), ("c", 3)])
print("From list of tuples:", d)
```

## Using zip() to Create Dictionary

```python
key = ["id", "name", "addr"]
values = ["s1", "Amit", "mumbai"]
d = dict(zip(key, values))
print("Dictionary from zip:", d)
```

## Accessing Dictionary Values

```python
student = {"name" : "Amit", "addr" : "mumbai", "course" : "AI"}

# Using square brackets (raises KeyError if key doesn't exist)
print("Using []:", student["name"])
# print(student["age"])  # Uncomment to see KeyError

# Using get() method (returns None if key doesn't exist)
print("Using get():", student.get("name"))
print("Using get() for missing key:", student.get("age"))
```

## Default Values with get()

```python
student = {"name" : "Amit", "addr" : "mumbai", "course" : "AI"}

# Providing default value if key not found
print("With default value:", student.get("age", "Not Found"))
```

## Modifying Dictionary Values

```python
# Direct assignment
student = {"name" : "Amit", "addr" : "mumbai", "course" : "AI"}
print("Original:", student)

student["name"] = "Rohit"
print("Modified:", student)
```

## Using update() Method

```python
student = {"name" : "Amit", "addr" : "mumbai", "course" : "AI"}
print("Before update:", student)

student.update({"name" : "Ravi"})
print("After update:", student)
```

## Key Methods Summary
- `dict[key]` - Access value (raises KeyError if missing)
- `dict.get(key, default)` - Access value (returns default if missing)
- `dict.update(other_dict)` - Update multiple keys
- `dict[key] = value` - Add or modify single key

```python
#Find the largest number in the list

n = [12, 45, 67, 23, 89, 34, 38, 90, 34,19, 34, 56, 78, 90, 23, 45]


largest = n[0]
for num in n:
    if num > largest:
        largest = num

print("The largest number is:", largest)

# Find number of even numbers in the list
count = 0
for num in n:
    if num % 2 == 0:
        count+=1
print("Total even numbers:", count)

#Find Duplicate numbers in the list

list = []
for num in n:
    if num not in list:
        list.append(num)

    print("Unique numbers:", list)

#Reverse a list using slicing
reversed_list = n[::-1]
print("Reversed list:", reversed_list)

```

```python
#Reverse a list without using reverse() method
reversed_list = []
for i in range(len(n)-1, 0, -1):
    reversed_list.append(n[i])
print("Reversed list:", reversed_list)
```

```python
#Calculate sum of the num in the list
total = 0
for num in n:
    total += num
print("Sum of all numbers:", total)
```

```python
#Second largest number in the list
num = n[0]
second_largest = n[0]
for num in n:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num
print("The second largest number is:", second_largest)
```
