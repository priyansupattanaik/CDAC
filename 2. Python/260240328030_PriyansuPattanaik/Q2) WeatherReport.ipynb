```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
days = np.array(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])
temperatures = np.array([25.5, 28.2, 22.8, 30.1, 27.3])
humidity = np.array([65, 70, 55, 80, 60])
rainfall = np.array([5.2, 0.0, 12.5, 2.3, 8.7])
wind_speed = np.array([12.5, 8.3, 15.2, 6.7, 10.1])
```

```python
print(f"Average Temp: {np.mean(temperatures):.2f}")
print(f"Highest Temp: {np.max(temperatures):.2f}")
print(f"Lowest Temp: {np.min(temperatures):.2f}")
print()
print(f"Average Humidity: {np.mean(humidity):.2f}")
print(f"Highest Humidity: {np.max(humidity):.2f}")
print(f"Lowest Humidity: {np.min(humidity):.2f}")
print()
print(f"Average Rainfall: {np.mean(rainfall):.2f}")
print(f"Highest Rainfall: {np.max(rainfall):.2f}")
print(f"Lowest Rainfall: {np.min(rainfall):.2f}")
print()
print(f"Average Wind Speed: {np.mean(wind_speed):.2f}")
print(f"Highest Wind Speed: {np.max(wind_speed):.2f}")
print(f"Lowest Wind Speed: {np.min(wind_speed):.2f}")
```

```python
total_rainfall = np.sum(rainfall)
print(f"Total Rainfall: {total_rainfall:.2f}")
```

```python
df = pd.DataFrame({
    'Day': days,
    'Temperature': temperatures,
    'Humidity': humidity,
    'Rainfall': rainfall,
    'Wind_Speed': wind_speed
})

print("Dataframe:",df)
```

```python
df['Average_Weather_Reading'] = (df['Temperature'] + df['Humidity'] + df['Wind_Speed']) / 3
df['Rainfall_Status'] = df['Rainfall'].apply(lambda x: 'Heavy' if x > 10 else 'Normal' if x > 5 else 'low')

print("Calculated Columns:")
print()print(df)
```

```python
high_temp_days = df[df['Temperature'] > 25]
print("Days with Temperature > 25:")
print(high_temp_days)

df_sorted_temp = df.sort_values('Temperature', ascending=False)
print("Sorted by Temperature:")
print(df_sorted_temp)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
bars = plt.bar(df['Day'], df['Rainfall'])
plt.title('Daily Rainfall Analysis')
plt.xlabel('Day')
plt.ylabel('Rainfall')
plt.grid(axis='y')


plt.subplot(1, 2, 2)
plt.plot(df['Day'], df['Temperature'], marker='o')
plt.title('Temperature Variation Over Days')
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.grid(True)
plt.ylim(min(df['Temperature']) - 2, max(df['Temperature']) + 2)


plt.tight_layout()
plt.show()
```
