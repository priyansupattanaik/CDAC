```python
#Create Dataframe
data = {
  "Name" : ["Amit", "Neha", "Raj"],
  "Marks" : [85, 92, 78]
}
df = pd.DataFrame(data)
df
```

```python
data = [["Amit", 85], ["Neha", 92], ["Raj", 78]]
df = pd.DataFrame(data, columns=['Name', 'Marks'])
df
```

```python
data = [["P", 23, "Mumbai"], ["K", 25, "Bhubaneswar"], ["L", 30, "Pune"]]
df = pd.DataFrame(data, columns=['Name', 'Age', 'City'])
df
```
