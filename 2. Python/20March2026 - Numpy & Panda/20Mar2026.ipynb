```python
#Grouping groupby()
data = {
    "Department" : ["HR", "IT", "HR", "IT", "HR", "IT"],
    "Salary" : [50000, 40000, 30000, 60000, 45000, 90000]
}

df = pd.DataFrame(data)
df.groupby("Department")["Salary"].mean()
```

```python
df1 = pd.DataFrame({"ID" : [1,2,3,4,5], "Name" : ["A", "B", "C", "D", "E"]})
df2 = pd.DataFrame({"ID" : [1,2,3,4,5], "Salary" : [10000, 20000, 30000, 40000, 50000]})

print(df1)
print()
print(df2)
```

```python
#Merge Operation
df3 = pd.merge(df1, df2, on="ID")
print(df3)
```
