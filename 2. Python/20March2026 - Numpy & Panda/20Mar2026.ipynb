**CASE STUDY 1: Student Performance Analysis**


```python
data = {
    "Id" : [1,2,3,4,5],
    "Name" : ["A", "B", "C", "D", "E"],
    "Marks" : [32, 33, 75, 45, 45]
}
df = pd.DataFrame(data)
df

#Display> 75 Marks
print(df[df["Marks"] > 40])

print()
#Add Grade to the students record
df["Grade"] = df["Marks"].apply(lambda x : "A" if x > 75 else "B")
print(df)
```

```python
import matplotlib.pyplot as plt

X = [1, 2, 3, 4, 5]
Y = [10, 20, 30, 40, 50]

plt.plot(X, Y)

plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
#Graph Decorating

import matplotlib.pyplot as plt

X = [1, 2, 3, 4, 5]
Y = [10, 20, 30, 40, 50]

plt.plot(X, Y, color = 'red', linestyle='--', marker='*')

plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt

X = [1, 2, 3, 4, 5]
Y = [10, 20, 30, 40, 50]
Y1 = [100, 200, 300, 400, 500]

plt.plot(X, Y1, label="Product A")
plt.plot(X, Y, label="Product B")
plt.legend() #This will show information

plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```
