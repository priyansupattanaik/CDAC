```python
import datetime

birthdate = datetime.date(2004, 5, 14)
today = datetime.date.today()
age_year = today.year - birthdate.year
age_days = today - birthdate
print(f"Age: {age_year} years")
print(f"Age: {age_days} days")
```

```python
import datetime

today = datetime.date.today()
deadline = datetime.date(2026, 4, 30)
if today < deadline:
    print("Project submission is on time.")
elif today == deadline:
    print ("Today is the deadline for project submission.")
else:
    print("Project submission is late.")
```

**REGEX**

Regular Expression:
----------------------
- It is a pattern matching technique use to
    - search text
    - validate input
    - extract data
    - replace the data

- Ex: 
    - Email Validation
    - Password Checking

module: import re
Functions: (methods)
-----------
re.match() : matches at beginning
re.search() : searches anywhere
re.findall() : returns all matches
re.sub : replace text
re.fullmatch() : matches entire string

Ex: re.search("abc", "abcdefgh")

. : Any character except
\d : Digits(0-9)
\D : Non-Digits
\w : word(a-z, A-Z, 0-9, _)
\W : Non-word
\s : Whitespace
\S : Non-Whitespace

Quntifiers:
------------
* : 0 or more
+ : 1 or more
? : 0 or 1
{n} : exactly n times
{n,m} : Between n and m times

Special Characters:
-------------------
` : OR
$ : End of string
[ ] : Character set
^ : Start of the string
() : use for grouping

Ex:
(r"\d+)
(r"[\w\.-]+@[\w+]\.")

r"[A-Za-z]+"

r"[6-9]\d{9}"



```python
import re

text = input("Enter String")

if re.fullmatch(r"[A-Za-z]+",text):
    print("Only alphabets !!!")
else:
    print("Not only characters")
```

```python
import re

text = input("Enter String")

if re.fullmatch(r"[6-9]\d{9}",text):
    print("Valid mobile number !!!")
else:
    print("Not a valid mobile number")
```
