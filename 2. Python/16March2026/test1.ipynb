# Object-Oriented Programming in Python
## Employee Management System Examples

## 1. Basic Variables Approach (Without OOP)

```python
# Traditional approach using separate variables
emp1 = 101
emp_name1 = "John Doe"
emp_salary1 = 50000.0

emp2 = 102
emp_name2 = "Jane Smith"
emp_salary2 = 60000.0

print("Employee 1:", emp1, emp_name1, emp_salary1)
print("Employee 2:", emp2, emp_name2, emp_salary2)
```

## 2. Basic Class with Method

```python
# Simple class demonstration
class Student:
    def display(self):  # self refers to the current object
        print("Getting Display")

s1 = Student()  # Object creation
s1.display()  # Calling the method using the object
```

## 3. Adding Attributes Dynamically

```python
# Adding attributes to objects dynamically
class Employee:
    pass

emp1 = Employee()  # First object
emp1.id = 101
emp1.name = "John Doe"
emp1.salary = 50000.0

emp2 = Employee()  # Second object (empty)

print("Employee 1 Details:")
print(f"ID: {emp1.id}")
print(f"Name: {emp1.name}")
print(f"Salary: {emp1.salary}")

print("\nEmployee 2 Details (No attributes assigned):")
try:
    print(f"ID: {emp2.id}")
    print(f"Name: {emp2.name}")
    print(f"Salary: {emp2.salary}")
except AttributeError as e:
    print(f"Error: {e} - Attributes not assigned to emp2")
```

## 4. Using Constructor (__init__ method)

```python
# Proper class with constructor
class Employee:
    def __init__(self, emp_id, emp_name, emp_salary):
        """Constructor method to initialize attributes"""
        self.emp_id = emp_id
        self.emp_name = emp_name
        self.emp_salary = emp_salary

    def show(self):
        """Method to display employee details"""
        print("Employee ID:", self.emp_id)
        print("Employee Name:", self.emp_name)
        print("Employee Salary:", self.emp_salary)
        print("-" * 30)

# Creating objects
emp1 = Employee(111, "John", 50000)
emp2 = Employee(112, "Jane", 60000)

# Displaying details
emp1.show()
emp2.show()
```

## 5. Method Calling Another Method

```python
# Class demonstrating method chaining
class Employee:
    def __init__(self, name):
        self.name = name
    
    def hello(self):
        print(f"Hello {self.name}!")

    def display(self):
        self.hello()  # Calling another method using self
        print(f"Employee Name: {self.name}")

# Creating and using object
emp = Employee("John Doe")
emp.display()
```

## 6. Multiple Objects with Constructor

```python
# Creating multiple objects with constructor
class Employee:
    def __init__(self, name):
        self.name = name

# Creating multiple objects
emp1 = Employee("John")
emp2 = Employee("Jane")

# Displaying names
print("Employee 1 name:", emp1.name)
print("Employee 2 name:", emp2.name)
```

## 7. Default Constructor Values

```python
# Class with default values in constructor
class Employee:
    def __init__(self):
        self.name = "No name provided!"
        self.salary = 0

    def display(self):
        print("Name:", self.name)
        print("Salary:", self.salary)
        print("-" * 20)

# Creating objects with default values
emp1 = Employee()
emp2 = Employee()
emp2.name = "Jane Smith"  # Modifying one object
emp2.salary = 60000

print("Employee 1 (Default):")
emp1.display()
print("Employee 2 (Modified):")
emp2.display()
```

## Summary of Key Concepts

1. **Class**: Blueprint for creating objects
2. **Object**: Instance of a class
3. **self**: Reference to the current object instance
4. **__init__**: Constructor method for initializing objects
5. **Attributes**: Variables that belong to an object
6. **Methods**: Functions that belong to a class
