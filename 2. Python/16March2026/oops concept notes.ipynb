# Object Oriented Programming (OOPs)

## What is OOP?
Object Oriented Programming is a programming paradigm that uses objects and classes. Instead of writing programs as procedures, object-oriented programming organizes software into objects that contain both data and behavior.

**Example:** Laptop - Real-world entity becomes an object in programming

## Advantages of OOP

1. **Code Reusability** - Classes can be reused using inheritance
2. **Modularity** - Programs can be divided into smaller parts
3. **Security** - Data hiding can be implemented using encapsulation
4. **Easy Maintenance** - Code becomes easy to update and manage
5. **Flexibility** - Supports polymorphism, allowing objects to be treated as instances of their parent class
6. **Real-world Modeling** - Programs resemble real-world objects which provide real-time solutions

### Syntax:

## Importance of 'self' Keyword

`self` is used to refer to the current instance (object) of the class.

1. **Refers to current object**
2. **Used to access instance variables** - Allows methods to access the variables of the object
3. **Used as Instance and Local variable**
4. **Used to call other methods of same class**
5. **Every object can have their own data**

**Key Points:**
- self represents the current object
- It can be used to access instance variables and methods
- The first parameter in instance method must be self
- It is automatically passed values when an object calls a method

## Types of Variables

### 1. Instance Variable
- Belongs to individual objects
- Each object has its own copy
- Defined using `self` keyword inside constructor or methods

**Key Points:**
- Created when object is created
- Stored inside the object memory
- Accessed using 'self' keyword
- Different objects may have different values

### 2. Static Variable (Class Variable)
- Belongs to the class
- Shared by all objects of the class
- Declared inside the class but outside the method

**Key Points:**
- One copy exists in the class
- All objects share the same value
- Accessed using class variable name

### 3. Local Variable
- Declared inside the method
- Accessible only within the method

**Key Points:**
- Temporary variable
- Exist only during method execution
- Not stored in object memory

### Example: Different Types of Variables

````python
# Example showing all types of variables
class Employee:
    # Static Variable (Class Variable)
    company = "CDAC"
    
    def __init__(self, name, salary):
        # Instance Variables
        self.name = name
        self.salary = salary
    
    def display(self):
        # Local Variable
        message = "Employee Details:"
        print(message)
        print(self.name, self.salary, Employee.company)

# Creating objects
e1 = Employee("John", 50000)
e2 = Employee("Jane", 60000)

# Displaying values
e1.display()
e2.display()

# Accessing class variable
print("\nCompany Name (accessed via class):", Employee.company)
````

## Types of Methods in OOP

### 1. Instance Method
- Operates on instance variables
- Requires object creation

**Key Points:**
- First Parameter must be 'self'
- Access instance variables
- Called using objects

**Example:** Car Management
- Car: object
- Instance variables: speed, fuel
- Instance method: accelerate(), brake(), speed()

### 2. Class Method
- Operates on class variables
- Defined using @classmethod decorator

**Key Points:**
- `cls` is the first parameter
- Used to modify class variables
- Called using class name

**Syntax:**
```python
@classmethod
def m1(cls):
    print("Hello")
```

### 3. Static Method
- Do not use instance variables or class variables
- Behave like normal functions placed inside the class

**Key Points:**
- @staticmethod decorator is used
- No 'self' or 'cls' parameter
- Used for utility functions

### Examples of Different Methods

````python
# Example 1: Static Method - Calculate area of cone
class Cone:
    @staticmethod
    def area(r, h):
        return 3.14 * r * r * h / 3

# Example 2: All types of methods
class Test:
    class_variable = 0
    
    def __init__(self, value):
        self.instance_variable = value  # Instance variable
    
    # Instance Method
    def instance_method(self):
        print(f"Instance method called. Instance variable: {self.instance_variable}")
    
    # Class Method
    @classmethod
    def class_method(cls):
        print(f"Class method called. Class variable: {cls.class_variable}")
    
    # Static Method
    @staticmethod
    def static_method(a, b):
        return a + b

# Using the methods
obj = Test(10)
obj.instance_method()  # Calling instance method
Test.class_method()     # Calling class method
result = Test.static_method(5, 10)  # Calling static method
print(f"Static method result: {result}")

# Using Cone static method
cone_area = Cone.area(5, 10)
print(f"\nCone area: {cone_area}")
````

## 1. Abstraction

Hiding the implementation details and showing only the functionality to the user. It helps to reduce complexity and increase efficiency.

### Abstract Class
An abstract class is a class that cannot be instantiated (cannot create objects) and is used as a blueprint for other classes.

**Contents of Abstract Class:**
- Abstract methods
- Normal methods
- Variables

**Key Points:**
- Defined using ABC (Abstract Base Class) module
- Must inherit from ABC
- May contain abstract methods
- Cannot create objects of abstract class
- Child classes must implement all abstract methods

### Abstract Method
An abstract method is a method that is declared but does not have an implementation in the abstract class. Child classes must override and implement this method.

### Syntax:

````python
from abc import ABC, abstractmethod

class AbstractClass(ABC):
    @abstractmethod
    def abstract_method_name(self):
        pass
````

### Example of Abstraction

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
    @abstractmethod
    def perimeter(self):
        pass
    
    def display_info(self):  # Normal method
        print("This is a shape")

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)

# Creating object of child class
rect = Rectangle(10, 5)
print(f"Rectangle Area: {rect.area()}")
print(f"Rectangle Perimeter: {rect.perimeter()}")
rect.display_info()

# This will give error - Cannot instantiate abstract class
# shape = Shape()  # TypeError
````

## 2. Encapsulation

Binding data (variables) and methods (functions) together in a single unit (class).

**Key Points:**
- Protecting sensitive data
- Preventing accidental modification
- Controlling access of variables using methods
- Improving maintainability of code

### Access Specifiers

| Type | Syntax | Description |
|------|--------|-------------|
| Private | `__var` | Within the class only |
| Protected | `_var` | Use in child classes |
| Public | `var` | Use anywhere |

### Rules to Implement Encapsulation
1. Sensitive Data should be private
2. Access should be given through setter/getter methods
3. Avoid Direct modification of important variables

### Example of Encapsulation

````python
class BankAccount:
    def __init__(self, account_number, balance, password):
        self.account_number = account_number  # Public variable
        self._balance = balance  # Protected variable
        self.__password = password  # Private variable
    
    # Getter method for balance
    def get_balance(self):
        return self._balance
    
    # Setter method for balance with validation
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print(f"Deposited {amount}. New balance: {self._balance}")
        else:
            print("Invalid amount")
    
    # Method to validate password (private variable access)
    def validate_password(self, input_password):
        return self.__password == input_password

# Creating account
account = BankAccount("ACC123", 5000, "secret123")

# Accessing public variable
print(f"Account Number: {account.account_number}")

# Accessing protected variable (should be avoided directly)
print(f"Balance (direct access): {account._balance}")

# Using getter method
print(f"Balance (via getter): {account.get_balance()}")

# Using setter method
account.deposit(1000)

# Accessing private variable (this will raise AttributeError)
# print(account.__password)  # Error

# Using method that accesses private variable
print(f"Password valid: {account.validate_password('secret123')}")
print(f"Password valid: {account.validate_password('wrong')}")
````

## Practice Examples

````python
# Example 1: Basic Employee Class
class Employee:
    def __init__(self, emp_id, emp_name, emp_salary):
        self.emp_id = emp_id
        self.emp_name = emp_name
        self.emp_salary = emp_salary

    def show(self):
        print("Employee ID:", self.emp_id)
        print("Employee Name:", self.emp_name)
        print("Employee Salary:", self.emp_salary)

emp1 = Employee(111, "John", 50000)
emp1.show()
print()
emp2 = Employee(112, "Jane", 60000)
emp2.show()
````

````python
# Example 2: Calling methods within class
class Employee9:
    def __init__(self, name):
        self.name = name
    
    def hello(self):
        print("Hello")

    def display(self):
        self.hello()  # display() calls hello() using self

emp = Employee9("John")
emp.display()
````

````python
# Example 3: Class with default values
class Employee:
    def __init__(self):
        self.name = "No name!"
        self.sal = 0

    def display(self):
        print("Name:", self.name)
        print("Salary:", self.sal)

emp1 = Employee()
emp1.display()
````

````python
# Example 4: Static Variable Example
class Employee:
    company = "CDAC"  # Static variable
    
    def __init__(self, name, sal):
        self.name = name  # Instance variable
        self.sal = sal    # Instance variable

    def display(self):
        print(self.name, self.sal, Employee.company)  # Accessing class variable

e1 = Employee("John", 50000)
e1.display()
e2 = Employee("Jane", 60000)
e2.display()
````
