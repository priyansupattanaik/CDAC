```python
#Adding attributes to the object uisng method

class Employee:
    def __init__(self, emp_id, emp_name, emp_salary): #constructor method #__init__ is a special method that is called when an object is created. 
                                                      #It is used to initialize the attributes of the object.
        self.emp_id = emp_id #self is used to refer to the current object and assign the value of emp_id to the emp_id attribute of the object.
        self.emp_name = emp_name
        self.emp_salary = emp_salary

    def show(self):
        print("Emp1 ID:", self.emp_id)
        print("Emp1 Name:", self.emp_name)
        print("Emp1 Salary:", self.emp_salary)

emp1 = Employee(111, "John", 50000) #object creation
emp1.show() #calling the method using the object
print()
emp2 = Employee(112, "Jane", 60000) #object creation
emp2.show() #calling the method using the object
```
