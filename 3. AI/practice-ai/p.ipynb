```python
#Concept: Norm

import numpy as np

A = np.array([2, 3])
print(np.linalg.norm(A)) # sqrt(13) ≈ 3.6055
```

```python
#L2 Norm

import numpy as np

v = np.array([3, -4])
print(np.linalg.norm(v))          # ord=2 by default                       # 5.0
```

```python
#L1 Norm
v = np.array([3, -4])
print(np.linalg.norm(v, ord=1))
```

```python
#L∞ Norm

v = np.array([3, -4])
print(np.linalg.norm(v, ord=np.inf))                   # 4
```

```python
#concept: Vector addition

A = np.array([2, 3])
B = np.array([2, 3])
print(A + B)   # [4, 6]
```

```python
#Concept: Scalar multiplication

S = 4
A = np.array([2, 3])
print(S * A)   # [8, 12]
```

```python
#Concept: Inner product

v1 = np.array([1, 2])
v2 = np.array([3, 4])
dot = np.dot(v1, v2)
print(dot)   # 11
```

```python
#Concept: Similarity

cos_sim = dot / (np.linalg.norm(v1) * np.linalg.norm(v2))
print("cosθ = ",cos_sim)   # 0.9839
```

```python
#Concept: Orthogonality

u = np.array([1, 0])
w = np.array([0, 1])
print(np.dot(u, w))          # 0 → orthogonal
```

```python
#Concept: Projection

proj_v1_on_v2 = (np.dot(v1, v2) / np.dot(v2, v2)) * v2
print(proj_v1_on_v2)   # [1.32, 1.76] #Used in PCA and regression.
```

```python
#Concept: PCA

data = np.array([[2.5, 2.4], [0.5, 0.7], [2.2, 2.9], [1.9, 2.2], [3.1, 3.0]])
mean = np.mean(data, axis=0)
centered = data - mean
cov = np.cov(centered.T)
eigvals, eigvecs = np.linalg.eig(cov)
print(eigvals)
```

```python
#Concept: SVD

data = np.array([[2.5, 2.4], [0.5, 0.7], [2.2, 2.9], [1.9, 2.2], [3.1, 3.0]])
U, S, Vt = np.linalg.svd(data)
print(f"Singular values: {S}")
```

```python
#Concept: Derivative

from sympy import symbols, diff

x = symbols('x')
f = x**2 + 3*x
df = diff(f, x)
print(df)   # 2*x + 3
```

```python
#Concept: Partial derivative

y = symbols('y')
g = x**2 * y + y**3
partial_x = diff(g, x)
partial_y = diff(g, y)
print(f"∂g/∂x = {partial_x}")   # 2*x*y
print(f"∂g/∂y = {partial_y}")   # x**2 + 3*y**2
```

```python
#Concept: Gradient

from sympy import Matrix

grad = Matrix([[diff(g, x)], [diff(g, y)]])
print(grad)
# Matrix([[2*x*y], [x**2 + 3*y**2]])
```

```python
#Concept: Gradient descent

def f_gd(x):
    return (x-3)**2 + 2

def df_gd(x):
    return 2*(x-3)

x0 = 10.0
lr = 0.1
for i in range(20):
    x0 -= lr * df_gd(x0)
print(f"Minimum found at x ≈ {x0:.4f}")   # ≈ 3.0000
```

```python
import sympy as sp

#Concept: Lagrange multipliers

lam = sp.symbols('lambda')
x_l, y_l = sp.symbols('x y')
f_l = x_l + y_l                # objective
g_l = x_l**2 + y_l**2 - 1      # constraint: unit circle
L = f_l - lam * g_l
eq1 = sp.diff(L, x_l)
eq2 = sp.diff(L, y_l)
eq3 = g_l
soln = sp.solve([eq1, eq2, eq3], [x_l, y_l, lam])
print("Stationary points:")
for s in soln[:2]:   # first two are real
    print(f"x = {s[0]}, y = {s[1]}, λ = {s[2]}")
```

```python
#Concept: Lagrange multipliers

import sympy as sp
x = sp.symbols('x')

lam = sp.symbols('lambda')
y= sp.symbols('y')
f = x**2 + y**2
g = x + y - 1
L = f - lam*g
eqs = [sp.diff(L,x), sp.diff(L,y), sp.diff(L, lam)]
sp.solve(eqs, (x, y, lam))
```
